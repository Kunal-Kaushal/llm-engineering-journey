```python
import os 
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display

GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
load_dotenv(override=True)
google_api_key = os.getenv('GOOGLE_API_KEY')
gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=google_api_key)
```

```python
response = gemini.chat.completions.create(model="gemini-2.5-flash-lite", messages=[{"role": "user", "content": "What is 2+2?"}])
print(response.choices[0].message.content)
```

```python
conversation = f"Charlie: Hi"
def alex():
    system_prompt = """
    You are Alex.
    Personality:
    - Highly argumentative and skeptical
    - Challenges every idea
    - Uses logic, sarcasm, and sharp reasoning
    - Enjoys pointing out flaws

    You are in a conversation with Blake and Charlie.
    """

    user_prompt = f"""
    Conversation so far:
    {conversation}

    Respond as Alex. Be critical, challenge assumptions, and disagree where possible.
    """
    return [{"role":"system","content":system_prompt },{"role":"user","content":user_prompt }]


def blake():
    system_prompt = """
    You are Blake.
    Personality:
    - Calm, balanced, and rational
    - Tries to mediate disagreements
    - Seeks practical solutions
    - Explains things clearly and neutrally

    You are in a conversation with Alex and Charlie.
    """

    user_prompt = f"""
    Conversation so far:
    {conversation}

    Respond as Blake. Try to balance the discussion and bring logical clarity.
    """
    return [{"role":"system","content":system_prompt },{"role":"user","content":user_prompt }]


def charlie():
    system_prompt = """
    You are Charlie.
    Personality:
    - Creative and optimistic
    - Thinks in ideas, possibilities, and innovation
    - Suggests new approaches and unconventional solutions
    - Sometimes ignores strict logic in favor of imagination

    You are in a conversation with Alex and Blake.
    """

    user_prompt = f"""
    Conversation so far:
    {conversation}

    Respond as Charlie. Be creative and propose new ideas or perspectives.
    """
    return [{"role":"system","content":system_prompt},{"role":"user","content":user_prompt }]
```

```python
def call_LLM(messages):
    response = gemini.chat.completions.create(model="gemini-2.5-flash-lite", messages=messages)
    return(response.choices[0].message.content)
```

```python

display(Markdown(f"{conversation} \n"))


for i in range(5):
    
    messages = alex()
    alex_response = call_LLM(messages)
    display(Markdown(f"Alex : {alex_response}"))
    conversation += f"\nAlex : {alex_response}"
    
    messages = blake()
    blake_response = call_LLM(messages)
    display(Markdown(f"Blake : {blake_response}"))
    conversation += f"\nBlake : {blake_response}"
    
    
    messages = charlie()
    charlie_response = call_LLM(messages)
    display(Markdown(f"Charlie : {charlie_response}"))
    conversation += f"\nCharlie : {charlie_response}"
    
    
```
