```python
gemini_model = "gemini-2.5-flash"
meta_model = "meta/llama-3.3-70b-instruct"
gemini_system = "You are a chatbot who is very argumentative; \
you disagree with anything in the conversation and you challenge everything, in a snarky way."

meta_system = "You are a very polite, courteous chatbot. You try to agree with \
everything the other person says, or find common ground. If the other person is argumentative, \
you try to calm them down and keep chatting."

gemini_messages = ["Hi there"]
meta_messages = ["Hi"]
```

```python
def call_gemini():
    messages = [{"role": "system", "content": gemini_system}]
    for g_msg, m_msg in zip(gemini_messages, meta_messages):
        messages.append({"role": "assistant", "content": g_msg})
        messages.append({"role": "user", "content": m_msg})
    response = gemini.chat.completions.create(model="gemini-2.5-flash", messages=messages)
    return response.choices[0].message.content
```

```python
call_gemini()
```

```python
def call_meta():
    messages = [{"role": "system", "content": meta_system}]
    for g_msg, m_msg in zip(gemini_messages, meta_messages):
        messages.append({"role": "user", "content": g_msg})
        messages.append({"role": "assistant", "content": m_msg})
    messages.append({"role": "user", "content": gemini_messages[-1]})
    response = nvidia.chat.completions.create(model=meta_model, messages=messages)
    return response.choices[0].message.content
```

```python
call_meta()
```

```python
call_gemini()
```

```python
gemini_messages = ["Hi there"]
meta_messages = ["Hi"]

display(Markdown(f"### Gemini:\n{gemini_messages[0]}\n"))
display(Markdown(f"### Meta:\n{meta_messages[0]}\n"))

for i in range(5):
    gemini_next = call_gemini()
    display(Markdown(f"### GEMINI:\n{gemini_next}\n"))
    gemini_messages.append(gemini_next)
    
    meta_next = call_meta()
    display(Markdown(f"### Meta:\n{meta_next}\n"))
    meta_messages.append(meta_next)
```
